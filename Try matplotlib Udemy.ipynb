```python
def auto_fig(rows=1, cols=1):
    fig,axes=plt.subplots(nrows=rows,ncols=cols)
    for row in range(rows):
        for col in range(cols):
            axes[row,col].plot(x,y)
            
auto_fig(5,3)
```
